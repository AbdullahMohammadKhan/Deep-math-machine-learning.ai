```python
import os, time, itertools, imageio, pickle
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.examples.tutorials.mnist import input_data
```

```python
def lrelu(x, th=0.2):
    return tf.maximum(th * x, x)

def generator(x, y, isTrain=True, reuse=False):
    
    with tf.variable_scope('generator', reuse=reuse):
        w_init = tf.contrib.layers.xavier_initializer()

        g_input = tf.concat([x, y], 1)

        fc1 = tf.nn.relu(tf.layers.dense(g_input, 128, kernel_initializer=w_init))
        fc2 = tf.layers.dense(fc1, 784, kernel_initializer=w_init)
        generated_outputs = tf.nn.tanh(fc2)

        return generated_outputs

def discriminator(x, isTrain=True, reuse=False):
    
    with tf.variable_scope('discriminator', reuse=reuse):
        
        w_init = tf.contrib.layers.xavier_initializer()

        fc1 = lrelu(tf.layers.dense(x, 128, kernel_initializer=w_init))
        source_logits = tf.layers.dense(fc1, 1, kernel_initializer=w_init)
        class_logits = tf.layers.dense(fc1, n_classes, kernel_initializer=w_init)

        return source_logits, class_logits
```

```python
x = tf.placeholder(tf.float32, shape=(None, 784))
y = tf.placeholder(tf.float32, shape=(None, 10))
z = tf.placeholder(tf.float32, shape=(None, 100))
isTrain = tf.placeholder(dtype=tf.bool)
```

```python
batch_size = 100
lr = 0.0002
train_epoch = 100
n_classes =10

mnist = input_data.read_data_sets("MNIST_data/", one_hot=True)
train_set = (mnist.train.images - 0.5) / 0.5
train_label = mnist.train.labels
```

```python
G_z = generator(z, y, isTrain)
D_real_source_logits, real_class_logits = discriminator(x,isTrain)
D_fake_source_logits, fake_class_logits = discriminator(G_z,isTrain, reuse=True)

D_loss_real_source = tf.reduce_mean(tf.nn.sigmoid_cross_entropy_with_logits(logits=D_real_source_logits, 
                                                                     labels=tf.ones([batch_size, 1])))
D_loss_fake_source = tf.reduce_mean(tf.nn.sigmoid_cross_entropy_with_logits(logits=D_fake_source_logits,
                                                                     labels=tf.zeros([batch_size, 1])))
loss_real_class = tf.reduce_mean(tf.nn.sigmoid_cross_entropy_with_logits(logits=real_class_logits, 
                                                                     labels=y))
loss_fake_class = tf.reduce_mean(tf.nn.sigmoid_cross_entropy_with_logits(logits=fake_class_logits,
                                                                     labels=y))

D_loss = D_loss_real_source + loss_real_class + D_loss_fake_source + loss_fake_class

G_loss_source = tf.reduce_mean(tf.nn.sigmoid_cross_entropy_with_logits(logits=D_fake_source_logits, 
                                                                labels=tf.ones([batch_size, 1])))

G_loss = G_loss_source + loss_real_class + loss_fake_class
```

```python
D_vars = tf.get_collection(tf.GraphKeys.TRAINABLE_VARIABLES, 'discriminator')
G_vars = tf.get_collection(tf.GraphKeys.TRAINABLE_VARIABLES, 'generator')

D_solver = tf.train.AdamOptimizer(learning_rate=1e-3,beta1=0.5).minimize(D_loss, var_list=D_vars)
G_solver = tf.train.AdamOptimizer(learning_rate=1e-3,beta1=0.5).minimize(G_loss, var_list=G_vars)
```

```python
sess = tf.InteractiveSession()
tf.global_variables_initializer().run()
```

```python
onehot = np.eye(10)

fixed_z_ = np.random.normal(0, 1, (25,100))
fixed_y_ = np.random.randint(0, 9, (25, 1))
fixed_y_ = onehot[fixed_y_.astype(np.int32)].squeeze()

def show_result(num_epoch):
    test_images = sess.run(G_z, {z: fixed_z_, y: fixed_y_, isTrain: False})


    size_figure_grid = 5
    fig, ax = plt.subplots(size_figure_grid, size_figure_grid, figsize=(5, 5))
    for i, j in itertools.product(range(size_figure_grid), range(size_figure_grid)):
        ax[i, j].get_xaxis().set_visible(False)
        ax[i, j].get_yaxis().set_visible(False)

    for k in range(size_figure_grid*size_figure_grid):
        i = k // size_figure_grid
        j = k % size_figure_grid
        ax[i, j].cla()
        ax[i, j].imshow(np.reshape(test_images[k], (28, 28)), cmap='gray')

    label = 'Epoch {0}'.format(num_epoch)
    fig.text(0.5, 0.04, label, ha='center')

    plt.show()
```

```python
for it in range(100):
        
    for iter in range(mnist.train.num_examples // batch_size):

        x_ = train_set[iter*batch_size:(iter + 1)*batch_size]
        y_ = train_label[iter*batch_size:(iter + 1)*batch_size]

        z_ = np.random.normal(0, 1, (batch_size, 100))
        
        _, D_loss_curr = sess.run([D_solver, D_loss], feed_dict={x: x_, y: y_, z: z_, isTrain: True})
        
        z_ = np.random.normal(0, 1, (batch_size, 100))
        _, G_loss_curr = sess.run([G_solver, G_loss], feed_dict={y: y_, z: z_,x: x_, isTrain: True})
        
    if it % 5 == 0 :
        
        show_result((it + 1))
        print('Iter: {}'.format(it))
        print('D loss: {:.4}'. format(D_loss_curr))
        print('G_loss: {:.4}'.format(G_loss_curr))
        print()
```
